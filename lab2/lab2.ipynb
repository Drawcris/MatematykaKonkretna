```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color

# Wariant zadania
wariant = 11
liczba_wierszy = wariant + 20  # 31
liczba_kolumn = wariant + 10   # 21

# Wczytaj obraz z lab1/11.webp
img = io.imread('../lab1/11.webp')
# Przekształć do skali szarości jeśli obraz jest kolorowy
if img.ndim == 3:
    img_gray = color.rgb2gray(img)
else:
    img_gray = img

# Wytnij fragment obrazu zgodnie z wariantem
X = np.array(img_gray[:liczba_wierszy, :liczba_kolumn])

# Rozkład SVD
U, S, VT = np.linalg.svd(X, full_matrices=False)

# Macierz korelacji kolumn (XTX)
corr_cols = X.T @ X
# Macierz korelacji wierszy (XXT)
corr_rows = X @ X.T

# Wizualizacja
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.title(f'Macierz korelacji kolumn (XTX), {liczba_kolumn} kolumn')
plt.imshow(corr_cols, cmap='hot')
plt.colorbar()

plt.subplot(1, 2, 2)
plt.title(f'Macierz korelacji wierszy (XXT), {liczba_wierszy} wierszy')
plt.imshow(corr_rows, cmap='hot')
plt.colorbar()
plt.show()

# Wnioski:
print(f'Użyto wariantu {wariant}: {liczba_wierszy} wierszy, {liczba_kolumn} kolumn. Macierze korelacji pokazują powiązania pomiędzy wierszami i kolumnami obrazu. Wartości bliskie zera oznaczają brak korelacji, wysokie wartości wskazują na silne powiązania.')
```
